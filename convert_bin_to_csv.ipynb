```python
import pandas as pd
from parser import DMU41Parser
from tqdm import tqdm

file_path = './results/20260626_202604_AlanVariance_success/dmu.bin'

with open(file_path, 'rb') as f:
    binary_data = f.read()

ax, ay, az, gx, gy, gz = [], [], [], [], [], []

parser = DMU41Parser()

for byte in tqdm(binary_data, desc="Parsing DMU data"):
    row = parser.parse_byte_essential(byte)
    if row is not None:
        ax.append(row['linear accelerations']['x'])
        ay.append(row['linear accelerations']['y'])
        az.append(row['linear accelerations']['z'])
        gx.append(row['angular rates']['x'])
        gy.append(row['angular rates']['y'])
        gz.append(row['angular rates']['z'])

df = pd.DataFrame({
    'ax': ax,
    'ay': ay,
    'az': az,
    'gx': gx,
    'gy': gy,
    'gz': gz
})

df.to_csv('./results/20260626_202604_AlanVariance_success/dmu.csv', index=False)
print("CSV saved")
```
